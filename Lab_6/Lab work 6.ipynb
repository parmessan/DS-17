### Завдання:

```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(0)

df = pd.DataFrame(data={'a':np.random.randint(0, 100, 30),
                        'b':np.random.randint(0, 100, 30),
                        'c':np.random.randint(0, 100, 30)})
df.head()
```

# _________________________________________________________________________________

Створити візуалізацію, аналогічно рисунку 
 - перші 3 графіки візуалізувати, викорстовуючи значення із df
 - останній це пряма пропорційність

```python
fig, ax = plt.subplots(2, 2, figsize = (13, 7))


ax[0][0].plot(df['a'])

ax[0][1].plot(df['b'])

ax[1][0].plot(df['c'])

ax[1][1].plot(df.index)

plt.tight_layout()
```

# _________________________________________________________________________________

Створити візуалізацію, аналогічно рисунку 

```python
fig, ax = plt.subplots(figsize = (13, 7))


ax.plot(df['a'])
```

# _________________________________________________________________________________

Створити візуалізацію, аналогічно рисунку

```python
fig, ax = plt.subplots(2, 1, figsize = (13, 7))


ax[0].plot(df['a'])

ax[1].plot(df['b'])

plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize = (13, 7))


ax.plot(df['a'])

ax.plot(df['b'])

plt.tight_layout()
```

# _________________________________________________________________________________

Створити візуалізацію, аналогічно рисунку 
- використовуйте колонки `a` та `b`
- задайте стиль 'darkgrid' за допомогою команди `sns.set_style`

```python
fig, ax = plt.subplots(figsize = (13, 7))

sns.set_style('darkgrid')

ax.plot(df['a'])

ax.plot(df['b'])

plt.tight_layout()
```

# _________________________________________________________________________________

Створити візуалізацію, аналогічно рисунку 
- для колонки `a` використайте червоний колір та лінію формату `-.`

- для колонки `b` використайте помаранчевий колір та товщину `10`

- для колонки `c` використайте жовтий колір та товщину `1` і маркер `o`

```python
fig, ax = plt.subplots(figsize = (13, 7))
sns.set_style('darkgrid')

ax.plot(df['a'], 'r-.')
ax.plot(df['b'], 'orange', linewidth = 10)
ax.plot(df['c'], 'yellow', marker = "o", linewidth = 1)
```

# _________________________________________________________________________________

Створити візуалізацію, аналогічно рисунку 
- і не забудьте про легенду :)

```python
fig, ax = plt.subplots(3, 1, figsize = (13, 7))

ax[0].plot(df["a"], label = "Line A")
ax[0].legend(loc = 4)
ax[1].plot(df["b"], label = "Line B")
ax[1].legend(loc = 6)
ax[2].plot(df["c"])
```

# _________________________________________________________________________________

Створити візуалізацію, аналогічно рисунку 
- використайте томатний колір та відстань між стовбцями 0.5

```python
fig, ax = plt.subplots(figsize = (13, 7))

ax.bar(df.index, df["a"], color = "tomato", width = 0.9, alpha = 0.5, label = "a")
ax.legend(loc = 1)
```

# _________________________________________________________________________________

Створити візуалізацію, аналогічно рисунку 
- добавте всі підписи та правильний маркер

```python
fig, ax = plt.subplots(figsize = (20, 10))

ax.plot(df.index, df["a"], marker = "^")

x_ticks = np.arange(0, 30, 1) 
plt.xticks(x_ticks)

ax.set_title("Info about 'a'", fontsize = 18)
ax.set_xlabel("Index", fontsize = 24)
ax.set_ylabel("Values", fontsize = 16)
```
